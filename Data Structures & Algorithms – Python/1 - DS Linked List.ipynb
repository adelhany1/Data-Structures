Lists is behind each other in memory,, Linked List is the nodes are spread out in different places in memory

Lists has indexes, LL don't

Linked list has a pointers (each node points to the next and last one points to None) and variables called head and tail

- append(add to the end-tail) in LL is O(1), doesn't matter the num of nodes is a list, num of operations to add node to the end is the same each time,,    in List: O(1)

- pop(remove from the end) in LL is O(n), because of the pointers start from the head through the list until the end (iterate through the entire list),,    in List: O(1)

- prepend(add to the beginning-head) in LL is O(1),,    in List: O(n)

- pop_first(remove from the beginning) in LL is O(1),,    in List: O(n)

- insert(add somewhere in the middle) of LL is O(n),,    in List: O(n)

- remove(from somewhere in the middle) of LL is O(n),,    in List: O(n)

pop is better for lists,, prepend and pop_first is better for the LL

Node consists of value + pointers and it's a dictionary when value = 4 (ex) and next = the next dict and so on until the final dict (final node) where the next will be = None

```python
#Class to Create a Node
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None


class LinkedList:
    def __init__(self, value):   # have to be initialized with only one value and then add values to it
        new_node = Node(value)
        self.head = new_node
        self.tail = new_node
        self.length = 1

    def print_list(self):
        temp = self.head
        while temp is not None:
            print(temp.value)
            temp = temp.next
            
   # Add item to the end
    def append(self, value):
        new_node = Node(value)
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node
        self.length += 1
        return True
    
   # Delete item from the end
    def pop(self):
        if self.length == 0:
            return None
        temp = self.head
        pre = self.head
        while(temp.next):
            pre = temp
            temp = temp.next
        self.tail = pre
        self.tail.next = None
        self.length -= 1
        if self.length == 0:
            self.head = None
            self.tail = None
        return temp  # return the node been removed ,, if you want the value make it temp.value (always want to return the entire node)

    
    # Add an item to the beginning of he linked list
    def prepend(self, value):
        new_node = Node(value)
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head      # make the new node point to the first item which is the head (attach them together)
            self.head = new_node           # Then make this new node the head itself
        self.length += 1
        return True  #optional , will use this method in new method that is why we need this True (requires poolean)
    
# Remove from the first
    def pop_first(self):
        if self.length == 0:
            return None
        temp = self.head  # make first node(want to remove) called temp
        self.head = self.head.next   # make the second node is the head 
        temp.next = None  # isolate it by make the next after it is None
        self.length -= 1
        if self.length == 0:     #After removing (if it was one item in LL)
            self.tail = None
        return temp  # return the node been removed ,, if you want just the value make it temp.value (always want to return the entire node)

    #retun the node at specific index
    def get(self, index):
        if index < 0 or index >= self.length:
            return None
        temp = self.head
        for _ in range(index):
            temp = temp.next
        return temp  # temp.value if you want the value
        
    def set_value(self, index, value):
        temp = self.get(index)
        if temp:  # because get method could return None if index is not right,, HERE it means if indes is coreect ...
            temp.value = value # if index is correct make its value equal to the new value
            return True # return True that the operation is DONE
        return False # return False if the index is not True
    
    def insert(self, index, value):
        if index < 0 or index > self.length:
            return False
        if index == 0:
            return self.prepend(value)
        if index == self.length:  # Last item in the LL
            return self.append(value)
        new_node = Node(value)  # If insert somewhere inbetween ..
        temp = self.get(index - 1)  # point to the right node to insert in   # insert new value before that index to be the new value in that index
        new_node.next = temp.next  # make the new node point to the next one after temp
        temp.next = new_node  # make the new node is the next according to temp ,, BY THERE we put the new between temp and the next one
        self.length += 1   
        return True  

    def remove(self, index):
        if index < 0 or index >= self.length:
            return None
        if index == 0:
            return self.pop_first()
        if index == self.length - 1:
            return self.pop()
        pre = self.get(index - 1)
        temp = pre.next # temp now is the one to remove
        pre.next = temp.next  # pass the 2 nexts to each other (that temp is in between them)
        temp.next = None   # isolate temp
        self.length -= 1
        return temp

    def reverse(self):
        temp = self.head
        self.head = self.tail
        self.tail = temp
        after = temp.next
        before = None
        for _ in range(self.length):
            after = temp.next
            temp.next = before
            before = temp
            temp = after
```

```python
my_linked_list = LinkedList(1)
my_linked_list.append(2)
my_linked_list.append(3)
my_linked_list.append(4)

my_linked_list.reverse()

my_linked_list.print_list()

```
