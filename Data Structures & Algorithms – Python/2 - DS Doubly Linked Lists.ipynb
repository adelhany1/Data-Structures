DLL have arrows that goes both ways right and left , Next and Previous

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None  # the right arrow (next)
        self.prev = None  # the left arrow (previous) , None because it's not pointing in anything yet
        

class DoublyLinkedList:
    def __init__(self, value):
        new_node = Node(value)  # First Node
        self.head = new_node
        self.tail = new_node
        self.length = 1

    def print_list(self):
        temp = self.head
        while temp is not None:
            print(temp.value)
            temp = temp.next
        
    def append(self, value):    # add to the end
        new_node = Node(value) 
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            new_node.prev = self.tail
            self.tail = new_node
        self.length += 1
        return True

    def pop(self):            # remove from the end
        if self.length == 0:
            return None
        temp = self.tail
        if self.length == 1:  # if ther's only one item in the DLL
            self.head = None
            self.tail = None 
        else:       
            self.tail = self.tail.prev  # make the prev to be the tail
            self.tail.next = None   # reomove the node after the previous (which became the tail (the end
            temp.prev = None    # Isloate the temp which is the final node that we removed
        self.length -= 1
        return temp

    def prepend(self, value):   # add node to the beggining
        new_node = Node(value)
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
        else:
            new_node.next = self.head  # make the next of the new point to head
            self.head.prev = new_node  # make the prev of the head point to new
            self.head = new_node      # change the head itself and make it the new
        self.length += 1
        return True

    def pop_first(self):         # Delete from the first
        if self.length == 0:
            return None
        temp = self.head   # the node we want to remove
        if self.length == 1:
            self.head = None
            self.tail = None
        else:
            self.head = self.head.next  # move the head to the next one
            self.head.prev = None       # make the prev of the head none (isolate the first one)
            temp.next = None         # Isolate it
        self.length -= 1
        return temp
    
    #retun the node at specific index
    def get(self, index):
        if index < 0 or index >= self.length:
            return None
        temp = self.head
        if index < self.length/2:
            for _ in range(index):
                temp = temp.next
        else:
            temp = self.tail
            for _ in range(self.length - 1, index, -1):
                temp = temp.prev  
        return temp
        
    def set_value(self, index, value):
        temp = self.get(index)
        if temp:
            temp.value = value
            return True
        return False
    
    def insert(self, index, value):
        if index < 0 or index > self.length:
            return False
        if index == 0:
            return self.prepend(value)
        if index == self.length:
            return self.append(value)

        new_node = Node(value)
        before = self.get(index - 1)
        after = before.next

        new_node.prev = before
        new_node.next = after
        before.next = new_node
        after.prev = new_node
        
        self.length += 1   
        return True  

    def remove(self, index):
        if index < 0 or index >= self.length:
            return None
        if index == 0:
            return self.pop_first()
        if index == self.length - 1:
            return self.pop()

        temp = self.get(index)
        
        temp.next.prev = temp.prev
        temp.prev.next = temp.next
        temp.next = None
        temp.prev = None

        self.length -= 1
        return temp
  
```
