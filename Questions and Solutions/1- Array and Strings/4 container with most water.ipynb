- Given n non-negative integers al, a2, ..., an , where each represents a point at coordinate (i, ai).
- n vertical lines are drawn such that the two endpoints of line i is at (i, ai) and (i, 0).
- Find two lines, which together with x-axis forms a container, such that the container contains the most water.
- n element: input array: 1,8,6,2,5,4,8,3,7 ..Building starts from the floor until these values(these values are on y-axis and x-axis is from 1 to 9)
- we want to find the container (biggest) using these lines, contain the most amount of water
- Best thing here is between the second line (8) and the last(7)
- the area that the fills is a rectangle, where the y-axis is the shortest building (line) between the 2 edges of them (7)
- SO , we want to Find the maximum difference between x position multiplied by the height (max area)
- How we determine which container is the best?

# Steps :
- Have 2 pointers at the ends of the array
  - I = 0    ,,, left pointer pointing to the first element
  - r = input length - 1   ,,, right pointer pointing to the last element
- loop as long as the left pointer is smaller than the right pointer
  - I = 0
  - r = input length - 1
  - maxarea = 0  ,,, initialize it at zero
  - while(I < r) {.....}
- Maximize the area between our 2 buildings
- Move the pointer pointing at the smallest building
  - while(I < r) {
  
    maxarea = max(maxarea, min(height[I] , height[r]) * (r-I))       }
    
    if(height[I] < height[r]):
    
       I += 1
    
    else:
    
    r -= 1

```python
class Solution:
    def maxArea(self, height):
        maxarea = 0
        l = 0
        r = len(height)-1  # index 8 (starting from 0 )

        while(l<r):   # indecies NOT Height
            maxarea = max(maxarea, min(height[l] , height[r]) * (r-l))
            if(height[l] < height[r]):
                l+=1
            else:
                r-=1
        return maxarea
```

```python
h = [1,8,6,2,5,4,8,3,7]
A = Solution()
A.maxArea(h)   # 7*7
```

- Time Complexity : O(N) because we did a single loop (single pass) over the input (container) using the 2 pointers
- Space Complexity : 0(1) ,, we didnt use external data structers, just the 2 pointers, didnt store anything into any data structure

```python
from typing import List
class Solutionn:
    def maxArea(self, height: List[int]) -> int:
        maxarea = 0
        l = 0
        r = len(height)-1

        while(l<r):
            maxarea = max(maxarea, min(height[l],height[r])*(r-l))
            if(height[l]<height[r]):
                l+=1
            else:
                r-=1
        return maxarea
```

```python
h = [1,8,6,2,5,4,8,3,7]
A = Solutionn()
A.maxArea(h)   # 7*7
```
