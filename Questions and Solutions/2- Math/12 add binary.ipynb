- Given 2 binary strings in form of binary numbers, return their sum
- Decimal numbers: 0 to 9
- Binary Nums : 0,1
- Decimal nums : 6 * 1 + 9 * 10 + 9 * 100 + 1 * 1000 = 1996 , each column represent a 10 to the power of the column position, starting from 0
- Same for binary numbers but multiples of 2 instead of 10 (2,4,8,16, ..)
- 1* 1 + 1 * 2 + 0 * 4 + 1 * 8  ,,, SO, 1011=11   (From right to left)

- Time complexity : O(N) , N is the maximum length of the 2 input strings, because we do a loop over the max length of the 2 input strings
- Space complexity : O(N) , Due to usage of an array to hold our result while we are looping

```python
class Solution(object):
    def addBinary(self, a, b):
        result = []
        carry = 0
        i = len(a)-1
        j = len(b)-1

        while i >= 0 or j >= 0 or carry:
            total = carry

            if i >= 0:
                total += int(a[i])
                i -= 1
            if j >= 0:
                total += int(b[j])
                j -= 1

            result.append(str(total % 2))
            carry = total//2
        return ''.join(reversed(result))
```

```python
a = Solution()
answer = a.addBinary("110","010")
print(answer)
```
